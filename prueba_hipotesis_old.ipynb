```python
from ipywidgets import interact
import ipywidgets as widgets

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random
import matplotlib.pyplot as plt
import math
from scipy.stats import norm


from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets
from ipywidgets import interact, interact_manual

mu_h0 = 68
mu_h1 = 70
medio=(mu_h0+mu_h1)/2
sigma = 0.6
n = 36
nsqrt = math.sqrt(n)
sigama_x = sigma/nsqrt
z1 =(mu_h0-mu_h1)/sigama_x
print(sigama_x)
print(z1)
fig, ax = plt.subplots()
desvios = 3.2
# for distribution curve
x0= np.arange(mu_h0-desvios*sigma,(mu_h0+desvios*sigma),0.001)
x1= np.arange(mu_h1-desvios*sigma,(mu_h1+desvios*sigma),0.001)
h0 = norm.pdf(x0, mu_h0, sigma)
print("......")
print(norm.pdf(166, mu_h0, sigma))
h1 = norm.pdf(x1, mu_h1, sigma)  
ax.plot(x0, h0)
ax.plot(x1, h1)
ax.set_title("Cumulative normal distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for fill_between
px=np.arange(mu_h0-desvios*sigma,medio,0.001)
h0aux = norm.pdf(px, mu_h1, sigma)
#ax.set_ylim(0,0.008)
ax.fill_between(px,h0aux,alpha=0.5, color='g')
# for text
#ax.text(-1,0.1,"cdf(x)", fontsize=20)
plt.show()
```
